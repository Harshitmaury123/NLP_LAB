```python
with open(r"C:\main\Desktop\NLP\tokens.txt", "r", encoding="utf-8") as f:
    text = f.read()

tokens = []
for word in text.split(','):
    if word:
        tokens.append(word)
tokens  
```

```python
tokens[0]
```

```python
def build(tokens):
    freq = {}
    for word in tokens:
        if word not in freq:
            freq[word] = 0
        freq[word] += 1
    return freq
```

```python
freq=build(tokens)
freq
```

```python
freq=sorted(freq.items(), key=lambda x: x[1], reverse=True)
freq
```

```python
import matplotlib.pyplot as plt

def plot_top(freq, n=100, title="शीर्ष 100 शब्द"):
    words, counts = zip(*freq[:n])
    plt.figure(figsize=(15,6))
    plt.bar(words, counts)
    plt.xticks(rotation=90, fontname="Nirmala UI", fontsize=10)  
    plt.title(title, fontname="Nirmala UI", fontsize=14)
    plt.show()
```

```python
plot_top(freq)
```

```python
stop_freq=[(w, c) for (w, c) in freq if c >= 517244]
stop_freq
```
